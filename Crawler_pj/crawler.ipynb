```python
import requests # 웹사이트에 접속해서 페이지 데이터를 가져오기 # import는 라이브러리를 사용하기 위해 쓰는 예약어 # requests는 웹페이지에서 HTML 데이터를 가져오는 라이브러리 
from bs4 import BeautifulSoup # 받아온 복잡한 HTML을 분석하기 # bs4는 BeautifulSoup4의 약자 BeautifulSoup4은 웹페이지의 데이터를 추출할 수 있도록 도와주는 분석 라이브러리
                              # 이후에 bs4.BeautifulSoup을 계속 입력하지 않기 위에 미리 선언

url = "https://www.yna.co.kr/news"  # 정보를 가져올 대상의 사이트 주소
headers = {"User-Agent": "Mozilla/5.0"}  # 웹사이트에 접속 거부를 방지하는 것 # Requests를 그냥 사용했을 때 로봇으로 판단해서 접속을 막는다 이때 User-Agent는 Mozilla 기반의 브라우저라고 인증하는 느낌
# 여기까지 완
# -----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

try:    # 아래의 코드를 실행하며 에러가 발생하면 멈추지 않고 except로 이동시키는 것
    response = requests.get(url, headers=headers)   # 웹사이트에 접속해서 받은 데이터를 response 변수로 지정 # requests를 통해 접근하고 .get 함수를 이용해 가져오기 get(데이터를 가져올 주소, )
    soup = BeautifulSoup(response.text, "html.parser")  # 받아온 데이터 response.text는 문자열 데이터 이를 파이썬이 이해할 수 있도록 Object형태로 변환

    print("--- [중복 제거 완료] 실시간 시사 뉴스 ---")

    all_tags = soup.find_all(['strong', 'a'])   # HTML 태그 중에서 제목이 주로 들어있는 <strong> 태그와 링크가 담긴 <a> 태그를 리스트로 만든다.

    # 중복을 체크하기 위한 빈 리스트를 만듭니다. (필터 역할)
    news_list = []  # 위에서 뽑아온 제목들을 하나씩 넣어 똑같은 제목이 나오면 걸러내기 위해 빈 리스트를 생성

    count = 0   # 뉴스의 번호를 매기기 위한 숫자 변수

    for tag in all_tags:    # 위에서 찾은 태그들을 하나씩 꺼내서 반복
        title = tag.get_text().strip()  # 태그 안에 있는 제목만 가져오고 앞뒤로 붙은 불필요한 공백은 제거

        # 필터링 조건:
        # 1. 글자 수가 20자 이상일 것
        # 2. 이미 출력한 뉴스(news_list)에 없을 것 (중복 제거 핵심!)
        if len(title) > 20 and title not in news_list:  # 제목이 너무 짧으면 광고일 수도 있어 20자가 넘는 진짜 뉴스 제목만 가져오는 조건
            news_list.append(title)     # 처음 나오는 제목이라면 리스트에 저장 다음에 또 나왔을 때 걸러내기 위함   
            count += 1
            print(f"{count}. {title}")
            print("-" * 50)

            if count >= 10: # 뉴스를 10개까지만 출력하고 반복문을 종료시킨다
                break

except Exception as e:  # 인터넷이 끊기거나 URL이 잘못되는 등 문제가 발생했을 때 프로그램을 종료시키는 것이 아닌 에러 메세지 출력
    print(f"에러 발생: {e}")
```
