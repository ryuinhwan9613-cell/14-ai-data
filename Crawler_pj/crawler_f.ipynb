프로젝트 이름 : 야구롤러

사용법 : 바로 실행시키면 됩니다.

실행을 시키면 url의 3페이지까지 해당되는 75개의 뉴스가 출력됩니다. 이때 중복되는 뉴스는 출력되지 않습니다.

```python
import requests                         # pip install requests
from bs4 import BeautifulSoup           # pip install beautifulsoup4

news_list = []
count = 0
print("--- 실시간 스포츠 야구 뉴스 ---\n")

for page in range(1,4):
    try:
        
        url = f"https://www.yna.co.kr/sports/baseball/{page}"
        response = requests.get(url)
        soup = BeautifulSoup(response.text, "html.parser")


        all_tags = soup.find_all(['strong', 'a'])
        

        for tag in all_tags:
            title = tag.get_text().strip()
            
            if len(title) > 10 and title not in news_list:
                news_list.append(title) 
                count += 1
                print(f"{count}. {title}")
                print("-" * 50)
                
                if count >= 75: 
                    break

    except Exception as e:
        print(f"에러 발생: {e}")
```
