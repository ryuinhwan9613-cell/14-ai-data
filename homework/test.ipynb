```python
n = int(input("숫자를 입력하세요"))

sum = 0
i = 1
process = ""

if  0 < n < 11:
    for i in range(1,n+1):

        sum += i
        if i == 1:
            process = "1"

        else:
            process = process + "+" + str(i)


    print(process,"=",sum)

else:
    print("양수만 입력하세요")
```

```python
def print_gugudan(dan,start=1,end=9):
    
    if start > end :
        print("잘못된 범위입니다")

    else :
        for i in range(start,end+1):
            print(f"{dan} X {i} = {dan*i}")

print_gugudan(5,3,8)
```
