### 문제 3: 1부터 N까지의 합
사용자가 입력한 숫자 N까지의 합을 계산하는 프로그램을 작성하세요.

**요구사항**:
- while 문 사용
- 양수만 입력 가능
- 계산 과정도 함께 출력

**실행 예시**:
```
숫자를 입력하세요: 5
1 + 2 + 3 + 4 + 5 = 15



````python
n = int(input("숫자를 입력하세요: "))

sum = 0
i = 1
process = "" 

if n > 0:
    for i in range(1,n+1):
        sum += i
        
        if i == 1:
            process = "1"

        else:
            process = process + "+" + str(i)

    print(process,"=",sum)

else:
    print("양수만 입력하세요")
````

### 문제 6: 숫자 피라미드
사용자가 입력한 높이만큼 숫자 피라미드를 출력하는 프로그램을 작성하세요.

**요구사항**:
- 중첩 for 문 사용
- 높이는 1~10 사이만 허용

**실행 예시** (높이 5 입력 시):
```
    1
   121
  12321
 1234321
123454321
```

````python
num = int(input("숫자를 입력하세요: "))

for i in range(1,num+1):
    print(" " * (num-i), end = "")

    for j in range(1, i + 1):
        print(j,end="")

    for j in range(i-1,0,-1):
        print(j,end="")
    
    print()
````

### 문제 7: 소수 판별기
사용자가 입력한 숫자가 소수인지 판별하고, 1부터 해당 숫자까지의 모든 소수를 출력하는 프로그램을 작성하세요.

**요구사항**:
- for 문과 if 문 사용
- 2 이상의 숫자만 입력 가능
- 소수 판별 로직 구현

**실행 예시**:
```
숫자를 입력하세요: 20
20은(는) 소수가 아닙니다.
 
1부터 20까지의 소수:
2, 3, 5, 7, 11, 13, 17, 19
총 8개의 소수가 있습니다.

````python
n = int(input("숫자를 입력하세요: "))
prime = 1

for i in range(2, n):

    if n % i == 0:
        prime = 0
        break

if prime == 0:
    print("소수가 아닙니다")


else:
    print("소수가 아닙니다")

primes = []

for num in range(2,n+1):
    count = 0
    for j in range(1,num+1):

        if num % j == 0:
            count += 1

    if count == 2:
        primes.append(num)

print(f"\n1부터 {n}까지의 소수: ")
print(str(primes)[1:-1])
print(f"총 {len(primes)}개의 소수가 있습니다.")


````
