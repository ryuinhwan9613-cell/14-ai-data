## 문제 2: 사칙연산 함수

두 숫자와 연산자를 입력받아 계산 결과를 반환(`return`)하는 함수를 작성하세요.

### 요구사항
- 함수 이름: `calculate`
- 매개변수: `a`, `b`, `operator`
- 지원 연산자: `+`, `-`, `*`, `/`
- 0으로 나눌 경우 `"0으로 나눌 수 없습니다"` 반환
- 잘못된 연산자 입력 시 `"잘못된 연산자입니다"` 반환

### 실행 예시
```python
print(calculate(10, 3, "+"))   # 13
print(calculate(10, 3, "-"))   # 7
print(calculate(10, 3, "*"))   # 30
print(calculate(10, 3, "/"))   # 3.3333333333333335
print(calculate(10, 0, "/"))   # 0으로 나눌 수 없습니다
print(calculate(10, 3, "%"))   # 잘못된 연산자입니다
```

def calculate(a, b, operator):
    
    if operator == "+":
        return a + b

    elif operator == "-":
        return a - b

    elif operator == "*":
        return a * b

    elif operator == "/":
        
        if b == 0:
            return "0으로 나눌 수 없습니다."  
        
        return a / b
    
    else:
        return "잘못된 연산자입니다."

## 문제 5: `*args` 활용 — 과목 점수 평균

여러 과목의 점수를 **개수 제한 없이** 입력받아 평균을 반환하는 함수를 작성하세요.

### 요구사항
- 함수 이름: `get_average`
- `*args`를 사용하여 가변 인수 처리
- 입력값이 없으면 `0` 반환
- 평균은 소수점 둘째 자리까지 반올림 # 둘째까지 나오게

### 실행 예시
```python   
print(get_average(80, 90, 100))        # 90.0
print(get_average(75, 82, 91, 68, 95)) # 82.2
print(get_average(100))                # 100.0
print(get_average())                   # 0
```

def get_average(*args):
    
    result = 0
     
    if args:
        for i in args:
        
            result = result + i

        return (result / len(args))

    else: 
        return 0

#### 다른 방법 (sum사용)

def get_average(*args):
    if len(args) == 0:
        return 0
    
    return round(sum(args)/len(args), 2)


---

## 문제 6: 구구단 출력 함수

구구단을 출력하는 함수를 작성하세요.

### 요구사항
- 함수 이름: `print_gugudan`
- 매개변수: `dan`, `start=1`, `end=9`
- `for`문으로 구구단 출력
- 잘못된 범위(`start > end`) 입력 시 `"잘못된 범위입니다"` 출력
- 반환값: 없음 (`print()`로 직접 출력)

### 실행 예시
```python
print_gugudan(5)
# 5 x 1 = 5
# 5 x 2 = 10
# ...
# 5 x 9 = 45

print_gugudan(3, 3, 7)
# 3 x 3 = 9
# 3 x 4 = 12
# 3 x 5 = 15
# 3 x 6 = 18
# 3 x 7 = 21

print_gugudan(7, 9, 2)
# 잘못된 범위입니다
```

def print_gugudan(dan, start = 1, end = 9):

    if start > end:
        print("잘못된 범위입니다.")
        # return 넣어야 함
    else: 
        for i in range(start, end + 1):
            print(f"{dan} X {i} = {dan * i}")

## 문제 13: 팩토리얼 함수

정수 `n`을 입력받아 `n!` 값을 반환하는 함수를 작성하세요.

### 요구사항
- 함수 이름: `factorial`
- 매개변수: `n`
- `for`문 사용
- `0! = 1` 처리

### 실행 예시
```python
print(factorial(5))  # 120
print(factorial(0))  # 1
```

def factorial(n):
    
    if n == 0:
        return 1
    
    result = 1
    
    for i in range(1, n + 1):
            result = result * i

    return result

- 매개변수: `text`, `ch`
- `count()` 메서드 사용 없이 직접 구현

### 실행 예시
```python
print(count_char("banana", "a"))  # 3
print(count_char("hello", "l"))   # 2
```

def count_char(text, ch):
    
    count = 0

    for i in text:

        if i == ch:
            count += 1

    return count
