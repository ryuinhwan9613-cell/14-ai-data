### 문제 7-3: BMI 계산기

키와 몸무게를 입력받아 BMI를 계산하고 상태를 출력하세요.

- BMI = 몸무게(kg) / (키(m) \*\* 2)
- 18.5 미만: 저체중
- 18.5 이상 23 미만: 정상
- 23 이상 25 미만: 과체중
- 25 이상: 비만

height = float(input("키를 입력하세요 (cm): "))
weight = float(input("몸무게를 입력하세요 (kg): "))

# 키를 미터로 변환

height_m = height / 100

# 여기에 코드를 작성하세요

height = float(input("키를 입력하세요 (cm): "))
weight = float(input("몸무게를 입력하세요 (kg): "))

height_m = height / 100
BMI = weight / (height_m ** 2)

if BMI < 18.5:
    print("저체중")

elif 18.5 <= BMI < 23:
    print("정상")

elif 23 <= BMI < 25:
    print("과체중")

else:
    print("비만")  

print(f"BMI 지수는 {BMI:.2f}입니다.")

### 윤년 판별기

연도를 입력받아 윤년인지 아닌지 판별하세요.

윤년 조건:

1. 4로 나누어떨어지는 해는 윤년
2. 단, 100으로 나누어떨어지는 해는 평년
3. 단, 400으로 나누어떨어지는 해는 윤년

year = int(input("연도를 입력하세요: "))
**예시:**

- 2020년: 윤년 (4로 나누어떨어짐)
- 2100년: 평년 (100으로 나누어떨어지지만 400으로는 안됨)
- 2000년: 윤년 (400으로 나누어떨어짐)

# 여기에 코드를 작성하세요

year = int(input("연도를 입력하세요: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("윤년입니다")

else:
    print("평년입니다.")
