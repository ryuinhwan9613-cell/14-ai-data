````
class Product:

    def __init__(self, product_id, name, price, weight, stock):
       
        self.product_id = product_id
        self.name = name
        self.price = price
        self.weight = weight
        self.stock = stock
        
    def get_info(self):
        return {
                "product_id" : self.product_id,
                "name" : self.name,
                "price" : self.price,
                "weight" : self.weight,
                "stock" : self.stock
               }

    def update_stock(self, quantity):
       
        if self.stock - quantity < 0:
            return False
        
        self.stock -= quantity
        return True

    def apply_discount(self, discount_rate):
        
        discount_price = self.price * (1 - discount_rate)
        return discount_price

laptop = Product("P001", "노트북", 1200000, 2.5, 10)
mouse = Product("P002", "무선마우스", 35000, 0.2, 50)
keyboard = Product("P003", "기계식키보드", 89000, 1.0, 30)
````

class ShoppingCart:
    """장바구니를 관리하는 클래스"""
    
    def __init__(self):
        """장바구니 초기화 (상품 목록을 저장할 리스트 생성)"""

    def add_product(self, product, quantity):

    def remove_product(self, product_id):

        pass
    
    def get_total_price(self):
        """장바구니 내 모든 상품의 총 금액 계산"""
        pass
    
    def get_total_weight(self):
        """장바구니 내 모든 상품의 총 무게 계산"""
        pass
    
    def show_cart(self):
        """
        장바구니 내용을 보기 좋게 출력
        각 상품의 이름, 수량, 가격 표시
        """
        pass

````
class ShoppingCart:

    def __init__(self):
        self.basket = []

    def add_product(self, product, quantity):
        basket = {
                "product" : product,
                "quantity" : quantity 
               }
        
        return self.basket.append(basket)
    
    def remove_product(self, product_id):
        
        for basket in self.basket:
            
            if basket["product"].product_id == product_id:
            
                self.basket.remove(basket)
                
                return True
            
        return False
    
    def get_total_price(self):
        
        total = 0

        for basket in self.basket:

            price = basket["product"].price
            i = basket["quantity"]
            total += (price * i)
        
        return total
    
    def get_total_weight(self):

        total_weight = 0

        for basket in self.basket:

            weight = basket["product"].weight
            i = basket["quantity"]
            total_weight += (weight * i)

        return total_weight
    
    def show_cart(self):

        if not self.basket:
            print("장바구니가 비어 있습니다.")
            return
        
        for basket in self.basket:
            a = basket["product"]
            b = basket["quantity"]

            print(f"상품명: {a.name}")
            print(f"수량: {b}개")
            print(f"합계: {a.price * b}원")

        print(f"총 결제 금액: {self.get_total_price()}원")


cart = ShoppingCart()
cart.add_product(laptop, 1)
cart.add_product(mouse, 2)
cart.add_product(keyboard, 1)
cart.show_cart()
````

        주문서 출력
        - 주문 ID
        - 고객명
        - 주문 상품 목록

````
class Order:

    def __init__(self, order_id, cart, customer_name):

        self.order_id = order_id
        self.cart = cart
        self.customer_name = customer_name

    def calculate_final_price(self):

        total_price = self.cart.get_total_price()
        total_weight = self.cart.get_total_weight()

        if total_weight >= 5.0:
            fee = 5000
        
        else:
            fee = 2500
        
        final_price = total_price + fee

        return final_price, fee
    
    def print_receipt(self):

        final_price, fee = self.calculate_final_price()
        
        print()
        print("주문서")
        print(f"주문 ID: {self.order_id}")
        print(f"고객 성함: {self.customer_name}")

        for item in self.cart.basket:
            c = item["product"]
            d = item["quantity"]
            print(f"{c.name} {d}개 {c.price * d}원")

        print(f"상품 금액: {self.cart.get_total_price()}원")
        print(f"배송비: {fee}원")
        print(f"최종 결제 금액: {final_price}원")

order = Order("ORD20260211001", cart, "홍길동")
order.print_receipt()        
````
