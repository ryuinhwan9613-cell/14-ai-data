- 초기 잔액: 10,000원
- 기능: 1.잔액조회 2.입금 3.출금 4.종료
- 출금 시 잔액 부족 체크

**실행 예시**:
```
=== ATM 시스템 ===
1. 잔액 조회
2. 입금
3. 출금
4. 종료
선택: 1
현재 잔액: 10,000원

선택: 2
입금액: 5000
입금 완료! 현재 잔액: 15,000원

선택: 3
출금액: 20000
잔액이 부족합니다. 현재 잔액: 15,000원

```

money = 10000

while True:
  
    print("=== ATM 시스템 ===")
    print("1. 잔액 조회")
    print("2. 입금")
    print("3. 출금")
    print("4. 종료")
    n = int(input("선택"))
    print(f"선택: {n}")
    
    if n < 1 or n > 4:

        print("1, 2, 3, 4의 숫자 중에서 입력해 주세요.")

    elif n == 1:
        
        print(f"현재 잔액: {money}\n")

    elif n == 2:
        
        money_add = int(input("입금액: "))
        money += money_add
        print(f"입금 완료! 현재 잔액: {money}\n")

    elif n == 3:
       
        money_sub = int(input("출금액: "))
        
        if money_sub > money:
            
            print(f"잔액이 부족합니다. 현재 잔액: {money}\n")

        else:    
           
            money -= money_sub
            print(f"출금 완료! 현재 잔액: {money}\n")

    elif n == 4:
        
        print("이용해 주셔서 감사합니다.")
