```python
#  파이썬 함수의 구조

# def 함수이름(매개변수1, 매개변수2, ...): # 매개변수 = 파라미터 # 매개변수의 값 = 인수
    # 수행할 문장 1
    # 수행할 문장 2
    # ...

def add(a, b):

    return a + b
```

```python
result = add(10, 20)
print(result)
```

```python
# 1. 일반적인 함수
# 입력값이 있고, 반환값이 있다

def add(a, b):
    result = a + b
    return result

sum = add(3, 4)
print(sum)
```

```python
# 2. 입력값이 없는 함수
# 입력값 X, 반환값 O

def say():
    return "hi"

a = say()
print(a)
```

```python
# 3. 반환값이 없는 함수
# 입력값 O , 반환값 X

def add(a, b):
    print(f"{a}, {b}의 합은 {a + b}입니다")

sum = add(3, 4)  # 호출을 해서 함수는 실행된다. 
print(sum)  # 반환이 없어서 값이 없음
```

```python
# 4. 입력값, 반환값 모두 없는 함수
# 입력값 X , 반환값 X

def say():
    print("hi")

a = say()
print(a)
```
