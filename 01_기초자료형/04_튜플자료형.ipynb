```python
#  튜플은 어떻게 만들까?

a = ()
b = (1,)  # 하나의 요소일 때만 마지막에 ,
c = (1, 2, 3)   # 여러 개는 마지막에 , 생략
d = 1, 2, 3    # 괄호 생략 가능
e = ("a", "b", ("ab", "cd"))
f = 1,
g = 1

print(type(a))
print(type(b))
print(type(c))
print(type(d))
print(type(e))
print(type(f))
print(type(g))
```

```python
# 튜플 요소 값을 삭제하려 할 때

a = (1, 2, "a", "b")

del a[0]
```

```python
# 튜플 요소 값을 변경하려 할 때

a = (1, 2, "a", "b")

a[0] = "c"
```

```python
a = (1, 2)
a = a * 2
print(a)
```

```python
a = 10
print(id(a))
a = 20
print(id(a)) 

a = [1, 2, 3]
print(id(a))

a.append(4)
print(a)
print(id(a))
```

```python
# 튜플 인덱싱 하기

a = (1, 2, 3, 4)
print(a[0])
```

```python
# 튜플 슬라이싱 하기

a = (1, 2, 3, 4)
print(a[:2])
```

```python
# 튜플의 더하기

a = (1, 2)
b = (3, 4)
c = a + b
print(c)
```

```python
# 튜플의 곱하기

a = (3, 4)
b = a * 3
print(b)
```

```python
# 튜플 길이 구하기

a = (1, 2, 3, 4)
print(len(a))
```
