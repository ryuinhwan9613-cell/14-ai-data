```python
# 큰따옴표 둘러싸기
"Life is too short, you need Python" # 문자열 자료형은 " " 안에 적어야 함 " " 없으면 변수로 생각 # ctrl + / 는 주석

# 작은따옴표 둘러싸기
'python is fun' 

# 큰따옴표 3개를 연속으로 둘러싸기 # 큰따옴표 3개 하면 줄 바꿈 가능
"""Life is too short, 
you need Python""" 

# 작은따옴표 3개를 연속으로 둘러싸기
b = '''Life is too short, 
you need Python'''

print(b)
```

```python
food = "Python's favorite food is perl" # 큰따옴표 대신에 작은따옴표 사용하면 에러
say = '"Python is very easy." he says' # shift + alt + 방향키 방향키로 복사

# 역슬래시를 사용해서 작은따옴표와 큰따옴표를 문자열에 포함하기
food = 'Python\'s favorite food is perl' # \는 뒤를 문자형으로 바꾸는 기능 이걸 이스케이프코드라고 함
food

say = "\"Python is very easy.\" he says" # alt + 방향키 방향키로 이동
```
