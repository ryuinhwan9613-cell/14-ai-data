```python
# for문의 기본 구조

# for 변수 in (리스트[] 또는 튜플() 또는 문자열):
#     수행할 문장 1
#     수행할 문장 2
#     ...

num_list = ["one", "two", "three"]

for num in num_list:
    print(num)
```

```python
a = [(1, 2), (3, 4), (5, 6)]    # 3개의 요소

for i in a:
    print(i)

```

```python
tuple_list = [(1, 2), (3, 4), (5, 6)]

for a in tuple_list:          # 언팩, 구조분해할당
    # print(first,second, first + second)
    print(a)
```

```python
# 언팩 (구조분해 할당)

a, b = (1, 2)
[a, b] = [1, 2]
[a, b] = (1, 2)
# [a, b] = (1, 2, 3) 불가능
[a, *b] = (1, 2, 3) # 가능


print(a)
print(b)
```

```python
marks = [90, 25, 67, 45, 80]
number = 0 

for mark in marks:
    number += 1

    if mark >= 60:
        print(f"{number}학생은 통과입니다.")
    
    else:
        print(f"{number}학생은 불합격입니다.")
```

```python
# for문과 continue문

marks = [90, 25, 67, 45, 80]
number = 0 

for mark in marks :
    number += 1

    if mark < 60 :
        continue

    else :
        print(f"{number}번 학생은 합격입니다용")
```

```python
# for문과 함께 자주 사용하는 range 함수

a = range(10)
print(a)

a = range(0, 10)
print(list(a))

a = range(1,10)
print(list(a))
```

```python
result = 0

for number in range(1, 10+1) :
    
    result += number
    print(result)

print(result)
```

```python
marks = [90, 25, 67, 45, 80]

for number in range(len(marks)):
    
    if marks[number] < 60 :
        continue
    
    print(f"{number + 1} 학생은 합격입니다.")
```

```python
# for와 range를 이용한 구구단 중첩 for문

for i in range(2, 10) :
    
    for j in range(1, 10) :
        
        print(i*j, end=" ")
    
    print("")
```

```python
# 출력 결과를 
# 2 X 1 = 2
# 2 X 2 = 4
# ...
# 2 X 9 = 18

# 3 x 1 = 3
# ...
# ...
# 9 X 9 = 81

for i in range(2, 10) :
    
    for j in range(1, 10) :
        
        print(f"{i} X {j} = {i * j}")

    print("")
```

```python
# *
# **
# ***
# ****
# *****

for i in range(1, 6) :
    
    print("*" * i)
    
```

```python
#     *
#    **
#   ***
#  ****
# *****

for i in range(1, 6):
    
    print(" " * (5 - i) + "*" * i)
```

```python
# 리스트 컴프리헨션 사용하기

a = [1, 2, 3, 4]
result = []

for num in a :
    
    result.append(num * 3)

print(result)

# 리스트 컴프리헨션 적용하기

a = [1, 2, 3, 4]

result = [num * 3 for num in a]
print(result)
```

```python
#  for문과 break문

for i in range(10) :
    
    if i == 5 :
        break
    
    print(i)
```

```python
# enumerate()함수 활용하기 (인덱스 값을 그냥 출력해줌)

fruits = ["apple", "banana", "orange"]

for i, fruit in enumerate(fruits) :
    
    print(f"{i + 1}번째 : {fruit}")

```

```python
# zip함수로 여러 리스트 함께 순회하기

names = ["홍길동", "유재석", "강호동"]
scores = [60, 70, 80]

for name, score in zip(names, scores) :

    print(f"{name}: {score}점")
```

```python
# 3개 이상의 리스트 - zip()

names = ["홍길동", "유재석", "강호동","류인환"]
korean = [60, 70, 80, 100000000000]
English = [80,  90, 70, 100000000000]

for name, score1, score2 in zip(names, korean, English) :
    
    print(f"{name}의 국어점수: {score1}, 영어점수: {score2}")
```
